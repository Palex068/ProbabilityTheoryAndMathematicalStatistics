```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from sklearn.linear_model import LinearRegression
```

__Задача 1__

```python
x1 = np.array([30,30,40, 40])
y1 = np.array([37, 47, 50, 60])

x2 = np.array([30,30,40, 40, 20, 20, 50, 50])
y2 = np.array([37, 47, 50, 60, 25, 35, 62, 72])

x3 = np.array([30,30,40, 40, 20, 20, 50, 50, 10, 10, 60, 60])
y3 = np.array([37, 47, 50, 60, 25, 35, 62, 72, 13, 23, 74, 84])
```

```python
r = round(np.corrcoef(x1, y1)[0][1], 2)

plt.scatter(x1, y1)

model = LinearRegression()
xr = x1.reshape(-1,1)
model.fit(xr, y1)
b0 = model.intercept_
b1 = model.coef_[0]

b0, b1
plt.plot(xr, xr * b1 + b0, color='red')

plt.title(f'r= {r}')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
r = round(np.corrcoef(x2, y2)[0][1], 2)

plt.scatter(x2, y2)

model = LinearRegression()
xr = x2.reshape(-1,1)
model.fit(xr, y2)
b0 = model.intercept_
b1 = model.coef_[0]

b0, b1
plt.plot(xr, xr * b1 + b0, color='red')

plt.title(f'r= {r}')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
r = round(np.corrcoef(x3, y3)[0][1], 2)

plt.scatter(x3, y3)

model = LinearRegression()
xr = x3.reshape(-1,1)
model.fit(xr, y3)
b0 = model.intercept_
b1 = model.coef_[0]

b0, b1
plt.plot(xr, xr * b1 + b0, color='red')

plt.title(f'r= {r}')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
models = [[x1, y1], [x2, y2], [x3, y3]]
```

```python
fig, axis = plt.subplots(3, 1)
for i in range(len(models)):
    cor = np.corrcoef(models[i])[0][1]
    axis[i].scatter(models[i][0], models[i][1])
    axis[i].set_title(f'r = {round(cor, 3)}')
    
plt.tight_layout()
```

```python
# reshape
for i in range(len(models)):
    models[i][0] = models[i][0].reshape(-1, 1)
```

```python
# regression
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

for i in range(len(models)):
    model = LinearRegression()
    model.fit(models[i][0], models[i][1])
    r_sq = model.score(models[i][0], models[i][1])
    print(f'model {i + 1}')
    print(f'R2:    {round(r_sq, 3)}')
    print(f'const: {round(model.intercept_, 3)}')
    print(f'beta:  {model.coef_}')
    print('-' * 30)
```

```python
# plot
fig, axis = plt.subplots(3, 1)
for i in range(len(models)):
    x = models[i][0]
    y = models[i][1]
    model = LinearRegression()
    model.fit(x, y)
    r_sq = model.score(x, y)
    const = model.intercept_
    beta = model.coef_[0]
    
    axis[i].scatter(x, y)
    axis[i].set_title(f'R2 = {round(r_sq, 3)}')
    
    axis[i].plot(x, beta * x + const, 'r')

plt.tight_layout()
```

__Задача 3__

```python
x = np.array([10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5])
y = np.array([8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68])
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

model = LinearRegression()
x = x.reshape(-1, 1)
model.fit(x, y)
r_sq = model.score(x, y)

const = model.intercept_
beta = model.coef_[0]
const, beta
```

```python
plt.scatter(x, y)
plt.plot(x, beta * x + const, 'g')
plt.title(f'R2 = {round(r_sq, 3)}')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

__Задача 4__

```python
# проверка остатков
from scipy import stats

y_hat = beta * x + const
y_hat = y_hat.reshape(1, -1)
resid = y - y_hat
print(y)
print(y_hat)
print(resid)

plt.figure(figsize=(3, 3))
plt.scatter(x, resid)
plt.xlabel('x')
plt.ylabel('residuals')
plt.show()

stats.shapiro(resid)
```

```python
# гомоскедастичность

plt.figure(figsize=(3, 3))
plt.scatter(y_hat, resid)
plt.xlabel('y_hat')
plt.ylabel('residals')
plt.show()
```
