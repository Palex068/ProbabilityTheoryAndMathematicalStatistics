## Задача 2
Провести двусторонний тест и ответить на вопрос, есть ли статистически значимые различия между средними $2$х нормально распределенных генеральных совокупностей, представленных следующими независимыми выборками:

```python
a = np.array([12, 10, 11, 19, 13, 11, 17, 15, 19, 14, 21, 18, 21, 11, 17, 14, 15, 17, 20, 19])
b = np.array([11, 13, 18, 15, 17, 18, 10, 21, 26, 15, 11, 12, 15, 17, 10, 18, 18, 12,21, 20])
```
Уровень статистической значимости принять за 5%

1. Используйте функцию в Python:<br>

```python
stats.ttest_ind(a, b, equal_var=True)
```

2. Имея p-value из функции рассчитать наблюдаемое значение критерия.

````python
a = np.array([12, 10, 11, 19, 13, 11, 17, 15, 19, 14, 21, 18, 21, 11, 17, 14, 15, 17, 20, 19])
b = np.array([11, 13, 18, 15, 17, 18, 10, 21, 26, 15, 11, 12, 15, 17, 10, 18, 18, 12,21, 20])
````

````python
r = np.corrcoef(a, b)
r
````

````python
plt.scatter(a, b)
plt.title(f"r = {round(r[0][1], 3)}")
plt.xlabel('a')
plt.ylabel('b')
plt.show
````

Решим двумя способами:
1. "Условно" в ручную
2. С помощью встроенной функции

````python
a_mean = np.mean(a) # Считаем среднее по a
b_mean = np.mean(b) # Считаем среднее по y
a_v = np.var(a, ddof=1) # Считаем несмещенную дисперсию по a
b_v = np.var(b, ddof=1) # Считаем несмещенную дисперсию по b
n_a = len(a) # количество элементов выборки a
n_b = len(b) # количество элементов выборки b
print(a_mean, b_mean, a_v, b_v, n_a, n_b)
result = [round(k, 4) for k in (a_mean, b_mean, a_v, b_v, n_a, n_b)]
result
````

````python
t_emp = (a_mean - b_mean) / np.sqrt(a_v / n_a + b_v / n_b)
t_emp
````

````python
alpha = 0.05
n = n_a

t1 = stats.t.ppf(alpha / 2, df=2 * (n - 1))
t2 = stats.t.ppf(1 - alpha / 2, df= 2 * (n - 1))
t1, t2 # находим критические значения
````

Из приведенных вычислений можем сделать вывод: <br>
$t_{наблюдаемое}$ не принадлежит критической области:

$-2.024 \leq -0.16 \leq 2.024 $, 

следовательно гипотеза $H_0$ не отвергается!

````python
stats.ttest_ind(a, b, equal_var=True)
````

Так как $pvalue > \alpha$, нулевая гипотеза принимается и статистически значимых различий между данными генеральными совокупностями нет. 

## Имея p-value из функции рассчитать наблюдаемое значение критерия.
